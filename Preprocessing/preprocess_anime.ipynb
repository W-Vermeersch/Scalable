```python
import pandas as pd
import kagglehub
import os
import numpy as np
```

```python
path = kagglehub.dataset_download("marlesson/myanimelist-dataset-animes-profiles-reviews")
```

```python
df = pd.read_csv(os.path.join(path, 'animes.csv'))
```

```python
df.rename(columns={'uid': 'anime_id'}, inplace=True)
```

```python
df.columns
```

```python
df = df[['anime_id', 'genre']].drop_duplicates()
df.groupby(['anime_id', 'genre']).value_counts()
```

```python
df.to_csv("../data/preprocessed/animes.csv")
```
