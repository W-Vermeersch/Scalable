```python
import pandas as pd
import kagglehub
import os
import numpy as np
```

```python
# fetch data from Kaggle if not already there
path = kagglehub.dataset_download("marlesson/myanimelist-dataset-animes-profiles-reviews")
```

```python
df = pd.read_csv(os.path.join(path, 'reviews.csv'))
```

```python
df.columns
```

```python
#do we remove also uid?
#Note: I added profile, because uid is not the id of the user but of the review itself
df = df[['uid', 'profile', 'anime_uid', 'score']].drop_duplicates()
```

```python
df.groupby(['profile', 'anime_uid']).value_counts()
```

```python
df = df.rename(columns={"anime_uid": "anime_id"})
df
```

```python
df = df[["profile", "anime_id","score"]]
df
```

```python
df.to_csv("../data/preprocessed/reviews.csv")
```
